```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
from tensorflow.keras.models import Sequential,load_model
```

```python
##load the trained model ,scaler,pickle,one hot encoder in ann.ipynb.
model=load_model('ann_churn_model.h5')
with open('scaler.pkl','rb') as f:
    scaler=pickle.load(f)
with open('label_encoder.pkl','rb') as f:
    label_encoder=pickle.load(f)
with open('onehot_encoder.pkl','rb') as f:
    onehot_encoder=pickle.load(f)
    
```

```python
#example data for prediction
data={'CreditScore':600,
      'Geography':'France',
        'Gender':'Male',
        'Age':40,
        'Tenure':3,
        'Balance':60000,
        'NumOfProducts':2,
        'HasCrCard':1,
        'IsActiveMember':1,
        'EstimatedSalary':50000
         }
```

```python
##categorical data encoding

geo_encoded=onehot_encoder.transform([['France']]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=onehot_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
#gender encoding
gender_encoded=label_encoder.transform([data['Gender']])
gender_encoded
```

```python
#combine all data into a single dataframe for prediction
input_df = pd.DataFrame([data]).drop("Geography", axis=1)
input_df['Gender'] = gender_encoded
input_df = pd.concat([input_df, geo_encoded_df], axis=1)
input_df
```

```python
#scaling the data
input_scaled=scaler.transform(input_df)
```

```python
# Make prediction
prediction = model.predict(input_scaled)
prediction_proba = prediction[0][0]
print(f'Churn Probability: {prediction_proba:.4f}')
if prediction_proba > 0.5:
    print('Prediction: Customer will churn')
else:
    print('Prediction: Customer will not churn')
```
